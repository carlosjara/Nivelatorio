```python
#¿Cúantos posibles "Job Title" existen en el Dataset?
dff_jobs = df["Job Title"].unique()
Cantidad = len(dff_jobs)
print("cantidad:", Cantidad)
```

```python
#¿Cúales son los "Employment Type" y en qué frecuencia se presentan en el Dataset?
dff_EmpType = df["Employment Type"]
dff_EmpType = dff_EmpType.value_counts()
dff_EmpType
```

```python
#¿Cuáles son los "Expertise Level" y "Experience Level" y cuál es el salario promedio general de cada uno?
dff_Level = df[["Expertise Level","Experience Level","Salary in USD"]]
dff_Level = dff_Level.groupby(["Expertise Level","Experience Level"]).agg({"Salary in USD":"mean"})
dff_Level
```

```python
#¿Cuáles son los "Salary Currency" y a qué tasa está covertido respecto al dolar cada uno?
dff_Rate = df[["Salary Currency","Salary","Salary in USD"]]
dff_Rate["Rate"] = dff_Rate["Salary"]/dff_Rate["Salary in USD"]
#dff_Rate= dff_Rate.groupby(["Salary Currency"]).agg({"Rate":"mean"})
dff_Rate
```

```python
#¿En cuántos paises diferentes se encuentran los empleados ("Employee Residence")?
dff_Residence = df["Employee Residence"]
dff_Residence = len(dff_Residence.value_counts())
print('Paises Diferentes: ',dff_Residence)
```

```python
#¿Cuál salario promedio por "Job Title" agrupado?
dff_JobSal = df[["Job Title","Salary in USD"]]
dff_JobSal = dff_JobSal.groupby(["Job Title"]).agg({"Salary in USD":"mean"})
dff_JobSal
```

```python
#¿Cuál es el salario promedio por país de residencia (Employee Residence)?
dff_ResidenceSal = df[["Employee Residence","Salary in USD"]]
dff_ResidenceSal = dff_ResidenceSal.groupby(["Employee Residence"]).agg({"Salary in USD":"mean"}).sort_values(by="Salary in USD", ascending=False)
#dff_ResidenceSal= dff_ResidenceSal.groupby("Employee Residence").mean().sort_values(by="Salary in USD", ascending=False)
dff_ResidenceSal
```

```python
#¿Cuál es el "Salary in USD" máximo y mínimo?¿A qué registro pertenece?
dff_ResidenceSal = df[["Salary in USD"]]
SalMax = df.nlargest(1, "Salary in USD")
SalMin = df.nsmallest(1, "Salary in USD")
SalMax
```

```python
#¿Cuál es el salario promedio (en USD) por rol por tamaño de compañía?
dff_JobCSal = df[["Company Size","Job Title","Salary in USD"]]
dff_JobCSal = dff_JobCSal.groupby(["Company Size","Job Title"]).agg({"Salary in USD":"mean"})
dff_JobCSal
```

```python
#muestre los datos agrupados por tamaño de la compañía, país y salario promedio.
dff_JobCloc = df[["Company Size","Company Location","Salary in USD"]]
dff_JobCloc = dff_JobCloc.groupby(["Company Size","Company Location"]).agg({"Salary in USD":"mean"})#.sort_values(by="Salary in USD", ascending=False)
dff_JobCloc
```
