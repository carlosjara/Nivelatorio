### Opciones para el conteo de categorías

#### Opción 1 - a mano ✊

```python
# Paso 1: Separar todas las categorías en una lista de listas
categorias_listas = df_movies["listed_in"].str.split(",").apply(lambda lista: [categoria.strip() for categoria in lista])
categorias_listas
```

```python
# Paso 2: Unir todas las listas en una sola lista
todas_categorias = [categoria for lista in categorias_listas for categoria in lista]
todas_categorias
```

```python
# Paso 3: Obtener los elementos únicos (usando un conjunto ---> no soporta repetidos)
elementos_unicos = set(todas_categorias)
elementos_unicos
```

```python
print(elementos_unicos, len(elementos_unicos))
```

#### Opción 2 - con la magia de la librería 🧙‍♂️

```python
# Paso 1: Separar todas las categorías en una única lista
listado_categorias = categorias_listas.explode()
listado_categorias
```

```python
# Paso 2: realizar el conteo
cantidad_por_categoria = listado_categorias.value_counts()
```

```python
import pandas as pd
from collections import Counter

# Ejemplo de DataFrame
data = {
    'show_id': [1, 2, 3, 4, 5],
    'type': ['Movie', 'Movie', 'Movie', 'TV Show', 'Movie'],
    'title': ['Movie 1', 'Movie 2', 'Movie 3', 'TV Show 1', 'Movie 4'],
    'director': ['Director 1', 'Director 2', 'Director 3', 'Director 4', 'Director 5'],
    'cast': ['Actor A, Actor B', 'Actor B, Actor C', 'Actor A, Actor C', 'Actor D, Actor E', 'Actor B, Actor E'],
    'country': ['Country 1, Country 2', 'Country 2', 'Country 1', 'Country 3', 'Country 2, Country 3']
}

df = pd.DataFrame(data)

# Paso 1: Convertir las celdas separadas por comas en registros individuales
df['country'] = df['country'].str.split(', ')

# Paso 2: Utilizar explode() en la columna "country"
df_exploded = df.explode('country')

# Paso 3: Agrupar por país y contar los actores
grouped = df_exploded.groupby('country')['cast'].apply(lambda x: ','.join(x)).str.split(',')
actor_counts_by_country = grouped.apply(lambda x: Counter(x))

# Paso 4: Obtener el actor con más repeticiones por país
most_common_actors_by_country = actor_counts_by_country.apply(lambda x: x.most_common(1)[0][0])

print(most_common_actors_by_country)
```
