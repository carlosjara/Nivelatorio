```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
def dropper(df):
    df.dropna(inplace=True)
    return df

def clean_headers(df, col_name='Order Date'):
    df = df.drop(df[df[col_name] == col_name].index)
    return df

def formatear_fecha(df,col_name='Order Date'):
    df['fecha_objeto'] = pd.to_datetime(df[col_name], format="%m/%d/%y %H:%M")

    df['nueva_fecha'] = df['fecha_objeto'].dt.strftime("%d/%m/%Y")
    df['nuevo_tiempo'] = df['fecha_objeto'].dt.strftime("%H:%M")

    df.drop(columns=['fecha_objeto'], inplace=True)
    return df
```

```python
%%time

ruta_carpeta = 'input/'  # TODO: Reemplaza esto con la ruta de tu carpeta

dataframes = []

for archivo in os.listdir(ruta_carpeta): # Recorre todos los archivos en la carpeta
    if archivo.endswith('.csv') or archivo.endswith('.xlsx'):
        ruta_archivo = os.path.join(ruta_carpeta, archivo)
        df = pd.read_csv(ruta_archivo) if archivo.endswith('.csv') else pd.read_excel(ruta_archivo)
        dataframes.append(df)
    
    #print(list(df.head(0).columns))
    #print(archivo,'\r\n',df.dtypes)
        
# Combinar todos los DataFrames en uno solo (** hablar del index)
df_combinado = pd.concat(dataframes, ignore_index=True)
# df_combinado
```

```python
%%time
df_combinado.pipe(dropper).pipe(clean_headers).pipe(formatear_fecha)
```
