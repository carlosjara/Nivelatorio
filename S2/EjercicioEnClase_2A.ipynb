### Variables categóricas versus numéricas

```python
df.head(1)
```

```python
sns.catplot(x="Experience Level", y="Salary in USD", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for Experience Level vs Salary in USD")
plt.show()
```

### Variables categóricas versus numéricas

```python
sns.catplot(x="Employment Type", y="Salary in USD", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for Employment Type vs Salary in USD")
plt.show()
```

### Variables categóricas versus numéricas

```python
sns.catplot(x="Company Size", y="Salary in USD", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for Company Size vs Salary in USD")
plt.show()
```

### Variables categóricas versus numéricas

```python
sns.catplot(x="Year", y="Salary in USD", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for Company Year vs Salary in USD")
plt.show()
```

### Variables categóricas versus numéricas

```python
sns.scatterplot(x="Year", y="Salary in USD", hue="Year", data=df, palette="Dark2", s=40)
plt.title("Relationship between proline, flavanoids and target")
plt.show()
```
